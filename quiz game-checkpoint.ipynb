```python
ans = input("Do you want to play ? (yes/no):").strip().lower()
if ans != "yes":
    print("Thank you ! See you next time.")
    exit()
print("Great! Let's get started.\n")
print("Here are some rules before you start the game.")
print("1. You will be asked 10 questions.")
print("2. Each correct answer will give you 3 points.")
print("3. Each wrong answer will deduct 2 points.")
print("4. You can skip a question by typing 'skip', but it will deduct 1 point.\n")
score = 0
questions = {
    "What is the capital of Bihar? ": "Patna",

    "What is 2 + 2? ": "4",

    "Which is the largest planet in our solar system? ": "jupiter",

    "Who wrote 'Hamlet'? ": "shakespeare",

    "What is the boiling point of water in Celsius? ": "100",

    "What is the chemical symbol for gold? ": "au",

    "Who painted the Mona Lisa? ": "da vinci",

    "What is the smallest prime number? ": "2",

    "Which is the hardest natural substance on Earth? ": "diamond",

    "What is the currency of Japan? ": "yen",

    "What is the tallest mountain in the world? ": "everest",

    "Who is known as the 'Father of Computers'? ": "babbage",

    "What is the main ingredient in guacamole? ": "avocado",

    "What year did the Titanic sink? ": "1912",

    "Which is the largest mammal in the world? ": "blue whale",

    "What is the fastest land animal? ": "cheetah",
    
     "How many continents are there on Earth? ": "7",

    "What is the smallest country in the world? ": "vatican city",

    "Who invented the light bulb? ": "edison",

    "What is the powerhouse of the cell? ": "mitochondria",

    "How many bones are there in the adult human body? ": "206",

    "Which is the longest river in the world? ": "nile",

    "Where is the Great Barrier Reef located? ": "australia",

    "When did World War II end? ": "1945",

    "How do fish breathe underwater? ": "gills",

    "Where is the Persian Gulf located? ": "middle east",

    "Which is the largest desert in the world? ": "sahara",

    "How many types of rocks are there? ": "3"
    
        
}

import random

selected_questions = random.sample(list(questions.items()), 10)

for question, answer in selected_questions:

    user_answer = input(question).strip().lower()

    if user_answer == "skip":

        score -= 1

        print("You chose to skip. 1 point deducted.")

    elif user_answer == answer:

        score += 3

        print("Correct! You earned 3 points.")

    else:

        score -= 2

        print(f"Wrong! The correct answer was '{answer}'. 2 points deducted.")

    print(f"Current score: {score}\n")

print(f"Your final score is: {score}")

print("Thank you for playing the quiz game!")
```
